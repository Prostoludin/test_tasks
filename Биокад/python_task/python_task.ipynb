#### 1. Изучите методы для сбора данных курсов валют в [API Центрального банка Российской Федерации](https://www.cbr.ru/development/SXML/) . Используя запросы к API, приведите значения заработной платы к рублям одним из вариантов: 1) простой вариант - на текущую дату  2) вариант посложнее - на дату создания (created).

```python
# преобразуем формат в дату
import datetime as dt
df['created'] = pd.to_datetime(df['created'])
df["date"] = pd.to_datetime(df['created'].dt.date)

# находим минимальные и максимальные даты для выгрузки курсов валют
min_date = df["date"].min()
max_date = df["date"].max()

# Преобразование дат в формат DD/MM/YYYY
min_date_formatted = min_date.strftime('%d/%m/%Y')
max_date_formatted = max_date.strftime('%d/%m/%Y')
```

```python
# Определим уникальные валюты используемые в выборке вакансий
df["currency"].unique()
```

Обнаружили, что заработную плату указывают в рублях, евро и долларах. Также есть пустые значения, которые считаем рублями.

```python
# Создадим словарь кодов необходимых для перевода в рубли валют
currency_dict = {
    "EUR": "R01239",
    "USD": "R01235"
}
```

```python
# Импортируем необходимые для получения и парсинга API ответа библиотеки
import requests
import xml.etree.ElementTree as ET

# Считываем и собираем в таблицу данные
data = []

for key, value in currency_dict.items():
    url = f"https://www.cbr.ru/scripts/XML_dynamic.asp?date_req1={min_date_formatted}&date_req2={max_date_formatted}&VAL_NM_RQ={value}"
    response = requests.get(url)

    if response.status_code == 200:
        root = ET.fromstring(response.content)

        for record in root.findall('.//Record'):
            date = record.get('Date')
            value = record.find('Value').text

            data.append({
                'Currency_name': key,
                'Date': pd.to_datetime(date),
                'Value': value.replace(',', '.')
            })
    else:
        print(f"Ошибка при выполнении запроса: {response.status_code}")

# Создание DataFrame из собранных данных
df_currencies = pd.DataFrame(data)
df_currencies["Value"] = df_currencies['Value'].astype(float)
df_currencies.info()
```

```python
# добавляем необходимые значения курса соответствующих валют в соответствующий день размещения вакансии
merged_df = pd.merge(df,
                     df_currencies, 
                     how = "left", 
                     left_on = ["date", "currency"], 
                     right_on = ["Date", "Currency_name"])
merged_df = merged_df.drop(columns = ["Currency_name", "Date"])

# заполняем пропуски в столбцах
merged_df["Value"] = merged_df["Value"].fillna(1)
merged_df["salary_to"] = merged_df["salary_to"].fillna(merged_df["salary_from"])
merged_df["salary_from"] = merged_df["salary_from"].fillna(merged_df["salary_to"])
merged_df["rur_salary_from"] = merged_df['salary_from']*merged_df["Value"]
merged_df["rur_salary_to"] = merged_df['salary_to']*merged_df["Value"]

# рассчитываем среднее значение зарплаты для каждой вакансии, потребуется в следующих заданиях
merged_df['rur_avg_salary'] = merged_df[['rur_salary_from', 'rur_salary_to']].mean(axis = 1)

# выводим результат
merged_df.head(10)
```

#### 3. Сделайте рейтинг (топ-30) навыков по всем вакансиям.

```python
# Импорт счетчика
from collections import Counter

# собираем все скилы в список
all_skills = df['skills'].str.split(';').explode().str.strip().to_list()

# удаляем пустые значения
all_skills = [skill for skill in all_skills if skill]

# считаем количество повторений
skill_counts = Counter(all_skills)

# содздаем датафрейм и сортируем его по количеству
skill_counts_df = pd.DataFrame(skill_counts.items(), columns=['Skill', 'Count'])
skill_counts_df = skill_counts_df.sort_values(by='Count', ascending=False)

# выводим результат
skill_counts_df.head(30)
```

```python
# фильтруем вакансии тестировщиков
qa_df = merged_df[merged_df['vacancy'].str.contains('QA', case=False)]

# считаем среднее значение заработной платы тестировщика по городам, сортируем по убыванию и выводим топ-10
avg_salary_top_rate = qa_df.pivot_table(values = 'rur_avg_salary', aggfunc= 'mean', index = 'area').sort_values('rur_avg_salary', ascending= False).head(10)

# выводим результат
avg_salary_top_rate
```

#### 5. Напишите функцию для поиска вакансий по переданным в нее значениям заработной платы, наличия тестового задания, наличия заданного текста в наименовании вакансии, наличия заданного навыка. В результате выведите количество найденных вакансий, среднюю заработную плату по найденным вакансиям и список ссылок найденных вакансий. 

```python
# функция для поиска вакансий

def vacancy_finder(salary, test_task, vacancy_part, skill):
    # Фильтрация по условиям зарплаты и наличия тестового задания
    df = merged_df.query('''has_test == @test_task and rur_salary_from <= @salary and rur_salary_to >= @salary''')

    # Фильтрация по подстроке в названии вакансии
    df = df[df['vacancy'].str.contains(vacancy_part, case=False, na=False)]

    # Фильтрация по подстроке в навыках
    df = df[df['skills'].str.contains(skill, case=False, na=False)]
    
    # количество вакансий
    vacancies_count = len(df)

    # среднее значение заработной платы по найденным вакансиям
    avg_salary = df['rur_avg_salary'].mean().round(2)

    # список ссылок на вакансии
    url_list = df['url'].to_list()

    return vacancies_count, avg_salary, url_list

# проверка функции
result = vacancy_finder(salary=400000, test_task=False, vacancy_part='Python', skill='Python')

print(result)
```

#### *Необязательно*. Усложненный вариант - сделать атрибуты функции опциональными, чтобы иметь возможность фильтрации вакансий по любой комбинации условий.

```python
# функция с фильтрацией по любой комбинации условий
def vacancy_finder(salary = None, test_task = None, vacancy_part = None, skill = None):
    # создаем копию исходного датафрейма
    df = merged_df.copy()
    
    # Фильтрация по условиям зарплаты 
    if salary is not None:
        df = df.query('''rur_salary_from <= @salary and rur_salary_to >= @salary''')

    # Фильтрация по наличию тестового задания
    if test_task is not None:
        df = df.query('''has_test == @test_task''')

    # Фильтрация по подстроке в названии вакансии
    if vacancy_part is not None:
        df = df[df['vacancy'].str.contains(vacancy_part, case=False, na=False)]

    # Фильтрация по подстроке в навыках
    if skill is not None:
        df = df[df['skills'].str.contains(skill, case=False, na=False)]
    
    # количество вакансий
    vacancies_count = len(df)

    # среднее значение заработной платы по найденным вакансиям
    avg_salary = df['rur_avg_salary'].mean().round(2)

    # список ссылок на вакансии
    url_list = df['url'].to_list()

    return vacancies_count, avg_salary, url_list

# проверка функции
result = vacancy_finder(salary=400000, test_task=False, vacancy_part='Python', skill='Python')

print(result)
```
